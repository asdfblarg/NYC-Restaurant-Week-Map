```python
import pandas as pd
import numpy as np
import requests
```

```python
BASE_URL = "https://program-api.nyctourism.com/restaurant-week"
HEADERS = {"x-api-key": "lTQSe929f34fohKaNq0OH53mdVL0yncvtqmuUG6i"}
```

```python
def make_request(page=1, lookup={}):
    payload = {
        "page": page,
        "lookup": lookup
    }
    response = requests.post(BASE_URL, headers=HEADERS, json=payload)
    return response.json()
```

```python
%%time
count = 0
page = 1
items = []

while True:
    print(f"Page {page}:", end="")
    res_json = make_request(page)
    items += res_json["items"]
    count += len(res_json["items"])
    page += 1
    print(f" {count} out of {res_json["count"]}")
#     time.sleep(0.5)
    if count >= res_json["count"]:
        break
        
lookup = res_json["lookup"]
```
