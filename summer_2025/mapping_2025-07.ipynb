```python
import folium
import pandas as pd
import numpy as np
```

```python
rw_resturants_df = pd.read_excel("rw_resturants_2025-07_full_v2.xlsx")
rw_resturants_df = rw_resturants_df.replace(np.nan, "")
rw_resturants_df
```

```python
len(rw_resturants_df[rw_resturants_df["latitude"].isna()])
```

```python
m = folium.Map(location=[40.7396, -73.9739], zoom_start=12)
for index, rest in rw_resturants_df.iterrows():
    if rest["latitude"] and rest["longitude"]:
        
        website_link = f'<a href="{rest.website}"target="_blank" style="font-size: 15px">Website</a></br>' if rest["website"] else ""
        pdf_link = f'<a href="{rest.pdf_url}"target="_blank" style="font-size: 15px">PDF</a></br>' if rest["pdf_url"] else ""
        
        tags = "; ".join(rest["tags"].split("; ")[4:]).replace("; ","</br><li>")
        if not rest["image_url"]:
#             rest["image_url"] = "https://nyc-tourism-public.s3.amazonaws.com/programs/25WRW-HP-Hero02-1920x1280.jpg"
            rest["image_url"] = "https://images.ctfassets.net/1aemqu6a6t65/6ueYLLIEDmwxJY95ZCnsoL/220bfc9c825e1110467f8638c1f61784/25-SRW-Hero-1920x1280.jpg"
            
        popup_html = (
            f'<a href="{rest.url}"target="_blank" style="font-size: 20px">{rest["name"]}</a></br>'
            f'<a href="{rest.url}" target="_blank"><img src="{rest["image_url"]}" width="350"></a>'
            f'{website_link}'
            f'{pdf_link}'
            f'{rest["meal_type"].replace("; ","</br>")}</br></br>'
            f'<li>{tags}'
        )

        folium.Marker(
            [rest["latitude"], rest["longitude"]],
            popup=popup_html,
            max_width=350,
            lazy=True,
            # tooltip=tooltip,
        ).add_to(m)

m
```

```python
rw_resturants_filtered_df = rw_resturants_df[rw_resturants_df["meal_type"].str.contains("30 Lunch", regex=False)]
map_30 = folium.Map(location=[40.7396, -73.9739], zoom_start=12)

for index, rest in rw_resturants_filtered_df.iterrows():
    if rest["latitude"] and rest["longitude"]:
        
        folium.Marker(
            [rest["latitude"], rest["longitude"]],
            popup=f'<a href="{rest.url}"">{rest["name"]}</>',
        ).add_to(map_30)

map_30
```
