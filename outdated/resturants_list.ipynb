```python
final_resturants_list = []

for prog_part in programParticipants:
    prog_part_data = prog_part['fields']["promocodeFor"]["fields"]
    
    name = prog_part_data["title"]
    url_slug = prog_part_data.get("slug", None)
    url = f'https://www.nyctourism.com/restaurant-week/{url_slug}' if url_slug else None
    address = prog_part_data["venueAddress"]
    website = prog_part_data["websiteUrl"]
    map_coords = prog_part_data.get("location", None)
    tags = [x["fields"]["title"] for x in prog_part_data["secondaryCategories"]]
    meal_type = prog_part['fields']["restaurantOfferTags"]
    
    pdf_url = None
    prog_part_pdf_assets = prog_part['fields'].get("restaurantPdfAsset")
    if prog_part_pdf_assets:
        if prog_part_pdf_assets[0].get("fields"):
            if prog_part_pdf_assets[0].get("fields").get("fileAttachment"):
                pdf_url = "https:"+prog_part_pdf_assets[0]["fields"]["fileAttachment"]["fields"]["file"]["url"]
                
    if prog_part_data.get("images"):
        img_url = f'http:{prog_part_data.get("images")[0].get("fields")["image"]["fields"]["file"]["url"]}'
    
    rest_dict = {
       "name": name,
        "url_slug": url_slug,
        "url": url,
        "img_url": img_url, 
        "pdf_url": pdf_url,
        "website": website,
        "tags": "; ".join(tags),
        "meal_type": "; ".join(meal_type),
        "address": address,
#         "map_coords": f'{map_coords["lat"]}, {map_coords["lon"]}' if map_coords else None,
        "latitude": map_coords["lat"] if map_coords else None,
        "longitude": map_coords["lon"] if map_coords else None,
    }
    
    final_resturants_list.append(rest_dict)
```

```python
rw_resturants_df = pd.DataFrame(final_resturants_list)
rw_resturants_df
```

```python
# rw_resturants_df.to_json("rw_resturants.json", orient="records")
# avoid seeing \/ instead of / for urls
with open("rw_resturants_v2.json", "w") as outfile:
    outfile.write(json.dumps(final_resturants_list))
```
